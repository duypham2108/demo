```python
from quickumls import QuickUMLS
```

```python
matcher = QuickUMLS("main")
```

```python
from pandas.io.json import json_normalize
import pandas as pd
```

```python
semantic_type = pd.read_table("SemanticTypes_2013AA.txt",sep="|",header=None)
semantic_types = pd.Series(semantic_type[2].values,index=semantic_type[1]).to_dict()
```

```python
def annotate(text):
    json = matcher.match(text, best_match=True, ignore_syntax=False)
    df = pd.DataFrame([], columns=["cui","end","ngram","preferred","semtypes","similarity","start","term"])
    for term in json:
        df = df.append(json_normalize(term).iloc[0,:])
    df = df.reset_index(drop=True)    
    df['type_meaning'] = df['semtypes'].apply(lambda x: list(x)[0]).map(semantic_types)
    return df
```
